```python
#Load the dataset
df=pd.read_csv("Amazon Sale Report.csv")
```

```python
#display the first 5 rows
print("Preview of data:")
display(df.head())
```

```python
#Check basic Info:
print("\n Dataset Info:")
df.info()
```

```python
#Basic stats
print("\n Summary Statistics:")
display(df.describe(include='all'))
```

```python
df.columns
```

```python
#1.Drop unwanted columns(we'll confirm which later)
df=df.drop(columns=['index','PendingS'],errors='ignore')
```

```python
#convert Date column to datetime
df['Date']=pd.to_datetime(df['Date'],format='%d-%m-%Y',errors='coerce')
```

```python
df['Date'].head(10)
```

```python
#3. Check for missing values
print("Missing Values:")
print(df.isnull().sum())
```

```python
#Check for duplicates
print("\n Number of duplicate rows:",df.duplicated().sum())
```

```python
#Preview after cleaning
df.head()
```

```python
#1.Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
#2. remove rows with missing Amount or Currency
df=df.dropna(subset=['Amount','currency'])
```

```python
df.shape
```

```python
#drop unnecessary column
df=df.drop(columns=['New'],errors='ignore')
```

```python
#4. drop rows missing shipping data
df=df.dropna(subset=['ship-city','ship-state','ship-postal-code','ship-country'])
```

```python
#check again
print("Data after cleaning:")
print(df.isnull().sum())

print("\n Remaining Rows:",len(df))
```

```python
import matplotlib.pyplot as plt
```

```python
#total revenue
total_revenue=df['Amount'].sum()
print(f"Total Revenue:{total_revenue:,.2f}")
```

```python
#Average order value
avg_order_value=df['Amount'].mean()
print(f"Average order value:{avg_order_value:,.2f}")
```

```python
#Total quantity sold
total_qty=df['Qty'].sum()
print(f"Total Quantity Sold:{total_qty:,}")
```

```python
#Orders by Status
order_status=df['Status'].value_counts()
print("\n Order Status distribution:")
print(order_status)
```

```python
#Montly sales trend
df['Month']=df['Date'].dt.to_period('M')
monthly_sales=df.groupby('Month')['Amount'].sum()
```

```python
plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line',marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.grid(True)
plt.show()
```

```python
#Top product categories by sales amount
top_categories=df.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(10)
print("Top 10 Categories by Sales Amount:")
print(top_categories)
```

```python
#Visualize it
top_categories.plot(kind='bar',figsize=(10,5),title='Top 10 Product Categories by Sales Amount')
```

```python
#Popular product sizes (if relevant)
top_sizes=df['Size'].value_counts().head(10)
print("\n Most popular product sizes:")
print(top_sizes)
```

```python
top_sizes.plot(kind='bar',figsize=(8,4),title='Top 10 Product Sizes')
```

```python
#average order value by category
avg_value_by_category=df.groupby('Category')['Amount'].mean().sort_values(ascending=False).head(10)
print("\n Top Categories by Average Order Value:")
print(avg_value_by_category)
```

```python
avg_value_by_category.plot(kind='bar',figsize=(10,5),title='Top 10 Categories by Avg Order Value')
```

```python
#Fulfillment method distribution
fulfilment_counts=df['Fulfilment'].value_counts()
print("Fulfilment Method Counts:")
print(fulfilment_counts)
```

```python
#Sales by fulfilment method
sales_by_fulfilment=df.groupby('Fulfilment')['Amount'].sum().sort_values(ascending=False)
print("\n Sales amount by Fulfilment method:")
print(sales_by_fulfilment)
```

```python
#Visualize
sales_by_fulfilment.plot(kind='bar',figsize=(8,4),title='Sales by Fulfilment Method')
```

```python
#average order value by fulfillment
avg_value_by_fulfilment=df.groupby('Fulfilment')['Amount'].mean().sort_values(ascending=False)
print("\n Average Order Value by Fulfilment:")
print(avg_value_by_fulfilment)
```

```python
avg_value_by_fulfilment.plot(kind='bar',figsize=(8,4),title='Avg Order Value by Fulfilment')
```

```python
#Top 10 states by total sales
top_states=df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)
print("Top 10 States by Sales Amount:")
print(top_states)
```

```python
top_states.plot(kind='bar',figsize=(10,5),title='Top 10 States by Sales Amount')
```

```python
#Top 10 cities by total sales
top_cities=df.groupby('ship-city')['Amount'].sum().sort_values(ascending=False).head(10)
print("\n Top 10 cities by sales Amount:")
print(top_cities)
```

```python
top_cities.plot(kind='bar',figsize=(10,5),title='Top 10 cities by sales amount')
```
